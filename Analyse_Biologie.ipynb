```python
import numpy as np 
import matplotlib.pyplot as plt 

heures=np.array([6,9,12,15,18,21])
temperature=np.array([12,16,22,26,23,18])

moyenne=np.mean(temperature)
max_temp=np.max(temperature)

print("Heures:",heures)
print("Température:",temperature)

print("Moyenne:",moyenne)
print("Max:",max_temp)

plt.plot(heures,temperature,marker='o')
plt.xlabel("heure")
plt.ylabel("temperature(°C)")
plt.title("évolution de la temperature le matin")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    "Etudiant":["Ali","Sara","Yacine","Nour","Lina","Omar"],
    "Math":[12,15,9,17,14,10],
    "Physique":[11,14,8,16,13,9],
    "Info":[13,16,10,18,15,11],
}

df=pd.DataFrame(data)

df["Moyenne"] = df[["Math","Physique","Info"]].mean(axis=1)

def mention(m):
    if m>=15:
        return"Très bien"
    elif m>=12:
        return"Bien"
    elif m>=10:
        return"passable"
    else:
        return"Ajourné"
df["Mention"]=df["Moyenne"].apply(mention)
print(df)
plt.bar(df["Etudiant"],df["Moyenne"])
plt.xlabel("Etudiant")
plt.ylabel("Moyenne")
plt.title("Moyenne des étudiants")
plt.show()
```
